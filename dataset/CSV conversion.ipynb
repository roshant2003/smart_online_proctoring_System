```python
import pandas as pd
import json

# Read the entire JSON file and load all JSON objects
with open('Med Dataset/train.json', 'r') as json_file:
    json_data = json_file.read()

# Split the file content into individual JSON objects
json_objects = json_data.strip().split('\n')

# Convert JSON objects to DataFrames and store them in a list
dataframes = []
for json_obj in json_objects:
    try:
        json_dict = json.loads(json_obj)
        df = pd.DataFrame([json_dict])
        dataframes.append(df)
    except json.JSONDecodeError:
        print(f"Skipping invalid JSON: {json_obj}")

# Concatenate all DataFrames into one
final_dataframe = pd.concat(dataframes, ignore_index=True)

# Save the combined DataFrame to a CSV file
final_dataframe.to_csv('output.csv', index=False)

print("Combined DataFrame saved as output.csv")
```
